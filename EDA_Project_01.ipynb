#### Price range in Airbnb

````python
airbnb_df['price'].describe()
````

````python
def price_range(price):
  ''' For different range of price we can categorize it into Cheap[minimun,100] , Affordable[100,450] , Expensive[450,maximun].
  In the argument pass the variable having range of values, it will give different category.
  '''
  
  if price<=100:

    return 'Cheap AirBnb'

  elif price>=100 and price<=450:

    return 'Affordable AirBnb'

  else:

    return 'Expensive AirBnb'
````

````python
sbn.countplot( x = airbnb_df['price'].map(price_range))

plt.xlabel('Range of price')
plt.ylabel('Count list')
plt.title('Variation of price with preference')
````

##### 2. What is/are the insight(s) found from the chart?

**From the above chart we can say that after ranging pricing from minimun to maximum value , most people are using average range of price  or below than that for accomodation and it is leading so high & very few people are taking expensive airbnbs. It is clear to say that two of one person wants affordable price rather than expensive price.**
